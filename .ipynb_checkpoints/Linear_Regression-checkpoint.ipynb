# Linear Regression using a Single Feature

##### smoker and age has the stronest correlation with the charges, we cant really use smokers cause it is yes or no value age can be used as it is continuos. let us try to estimate the charges usong age for non smokers

```python
non_smoker_df = medical_df[medical_df.smoker=='no']
```

```python
plt.title('Age vs Charges')
sns.scatterplot(data=non_smoker_df, x='age', y='charges', alpha =0.7, s=15);
```

# MODEL 
charges=w*age+b

w and b are called the weight and bias

```python
def estimated_charges(age ,w ,b):
    return w * age + b
```

```python
ages = non_smoker_df.age
ages
```

```python
estimated_charge = estimated_charges(ages, 50, 100)
```

```python
estimated_charge
```

```python
plt.plot(ages, estimated_charge, 'r-o')
plt.xlabel('Age')
plt.ylabel('Estimated Charges')
```

##### checking if our model fits the data

```python
target = non_smoker_df.charges
```

```python
plt.plot(ages, estimated_charge,'r',alpha=0.9)
plt.scatter(ages,target,s=8,alpha=0.8)
plt.xlabel('Age')
plt.ylabel('Charges')
plt.legend(['Estimate','Actual'])
```

##### the rate needs to be higher as we go from 30 to 40 the increase needs to be higher the bias looks good at this point

```python
def try_parameters(w, b):
    ages = non_smoker_df.age
    target = non_smoker_df.charges

    estimated_charge =  estimated_charges(ages, w, b)

    plt.plot(ages, estimated_charge, 'r', alpha=0.9)
    plt.scatter(ages, target, s=8, alpha=0.9)
    plt.xlabel('Age')
    plt.ylabel('Charges')
    plt.legend(['Estimate', 'Actual'])
```

```python
try_parameters(60, 200)
```

```python
try_parameters(400, -8600)
```

1. a way to measure numerically how well the lines fits the points
2. once measure of fit is computed we need to modify w and b for best fit line, strating from random guess

```python
targets = non_smoker_df.charges
targets
```

```python
predictions = estimated_charge
predictions
```

```python
import numpy as np
```

```python
def rmse(target, prediction):
    return np.sqrt(np.mean(np.square(targets - prediction)))
```

```python
predicted = estimated_charges(non_smoker_df.age,50, 100)
```

```python
rmse(targets,predicted)
```

On average each number in the prediction differs by $8461, this is also cslled the loss

```python
def try_parameters(w, b):
    ages = non_smoker_df.age
    target = non_smoker_df.charges
   
    predictions =  estimated_charges(ages, w, b)
    
    plt.plot(ages, predictions, 'r', alpha=0.9)
    plt.scatter(ages, target, s=8, alpha=0.9)
    plt.xlabel('Age')
    plt.ylabel('Charges')
    plt.legend(['Prediction', 'Actual'])
   
    loss = rmse(target,predictions)
    print("RMSE Loss", loss)
```

```python
try_parameters(350,-5500)
```

# Model Improvements 

To compare the importance of each feature in the model, our firts instinct might br to compre their weights

```python
model.coef_
```

```python
model.intercept_
```

```python
weights_df = pd.DataFrame({
    'features':np.append(input_cols, 1),
    'weight' : np.append(model.coef_, model.intercept_)
})
weights_df
```

While it seems like BMI and the 'northeast' have a higher weight than age, keep in mid thta the range of values for BMI is limited (15 to 40) and the northeast column only takes the value 0 and 1

beacause different column had different ranges, we run into two issues:
1. we cant compare the weights of different column to identify which features are important
2. A column with a large range of inputs may disproportionately affect the loss and dominate the optmizaton process

For this reason, it is a common practice to scale(or standardize) the values in numeric column by subtracting the mean and deviding it by the standard deviation

```python
medical_df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numeric_cols = ['age','bmi', 'children']
scaler = StandardScaler()
scaler.fit(medical_df[numeric_cols])
```

```python
scaler.mean_
```

```python
scaler.var_
```

we can scale the data as follows

```python
medical_df[numeric_cols]
```

```python
scaled_inputs = scaler.transform(medical_df[numeric_cols])
scaled_inputs
```

```python
cat_cols = ['smoker_code','sex_code','northeast', 'northwest', 'southeast', 'southwest']
categorical_data = medical_df[cat_cols].values
```

```python
inputs
```

```python
#create inputs and targets
inputs, targets = np.concatenate((scaled_inputs,categorical_data),axis=1), medical_df[['charges']]

#create and train the model
model =  LinearRegression().fit(inputs, targets)

#Genrate predicitons
predictions = model.predict(inputs)

#compute loss to evaluate the model
loss = rmse(targets, predictions)
print('Loss:', loss)
```

```python
weights_df = pd.DataFrame({
    'features':np.append(numeric_cols + cat_cols, 1),
    'weight' : np.append(model.coef_, model.intercept_)
})
weights_df.sort_values('weight', ascending=False)
```

As we can see, the most important features are smoker, age, BMI
