## Modeling - baseline

In this case, if our baseline model is just randomly guessing the species of flower, or guessing a singlr species for every data point, we would expect to have a model accuracy of 0.33 or 33%, since we have 3 different classes that are evenly balanced

So our model should at least beat 33% accuracy

## Modeling - Simple manual model

Let's manually look at our data and decide some cutoff points for classfication

```python
def single_feature_prediction(petal_length):
    """ Predicts the iris species given the petal length"""
    if petal_length < 2.5:
        return 0  #return "setosa"
    elif petal_length < 4.8:
        return 1  #return "versicolor"
    else:
        return 2  #return "virginica"
```

```python
df_train['petal length (cm)'].values
```

```python
x_train[:, 2]
```

```python
manual_y_predictions =np.array([ single_feature_prediction(val) for val in x_train[:, 2]])
```

```python
y_train
```

```python
manual_y_predictions == y_train
#will print true when the prediction is correct
```

```python
manual_model_accuracy=np.mean(manual_y_predictions == y_train)
print(f"Manual model accuarcy : {manual_model_accuracy *100:.2f}%")
```

```python
#fig, axs = plt.subplots(2,2, figsize=(10,10))
```

```python
def plot_incorrect_predictions(df_predictions, x_axis_feature, y_axis_feature):
    fig, axs = plt.subplots(2,2, figsize=(10,10))
    axs = axs.flatten()
    sns.scatterplot(x = x_axis_feature, y=y_axis_feature, hue="prediction_label", data = df_predictions, ax=axs[0])
    sns.scatterplot(x = x_axis_feature, y=y_axis_feature, hue="target_name", data = df_predictions, ax=axs[1])
    sns.scatterplot(x = x_axis_feature, y=y_axis_feature, hue="correct_prediction", data = df_predictions, ax=axs[2])
    axs[3].set_visible(False)

    plt.show()
```

```python
plot_incorrect_predictions(df_predictions,"petal length (cm)","petal width (cm)")
```

## Model Tuning

```python
for reg_param in (1,1.3,2,5,10):
    print(reg_param)
    model = LogisticRegression(max_iter=200,C=reg_param) #basian hyper parameter optimization can also be used
    accuracies = cross_val_score(model, x_train, y_train, cv=5, scoring="accuracy")
    print(f"accuarcy : {np.mean(accuracies) *100:.2f}%")
```

## Final Model

```python
model = LogisticRegression(max_iter=200,C=2)
```
